```python
import pandas as pd
```

```python
pred = pd.read_csv("historical_model/predictions/senate_pred.csv").set_index("District").drop_duplicates()
print(pred)
```

```python
real = pd.read_csv("historical_model/test_data/merged_senate_districts_2015_2016.csv")
real = real[['District','sex', 'party']].drop_duplicates().set_index("District")
real['sex'] = real['sex'].fillna(1)
real['sex'].replace('f', 1, inplace=True)
real['sex'].replace('m', 0, inplace=True)
real['sex'] = real['sex'].astype(int)
real['party'].replace('Democratic', 1, inplace=True)
real['party'].replace('Republican', 0, inplace=True)
real['party'] = real['party'].fillna(real['party'].mean())

real['female_dem'] = 0
for index, row in real.iterrows():
    if row.sex == 1 and row.party == 1:
        real.set_value(index, 'female_dem', 1)
print(real)
```

```python
female_real = list(real['female_dem'])
female_pred = list(pred['Probability_Female_Dem'].apply(lambda x: round(x)))
print(len(female_real), len(female_pred))

count = 0
for i in range(len(female_real)):
    if female_real[i] == female_pred[i]:
        count+=1
print(count/len(female_real))
```
