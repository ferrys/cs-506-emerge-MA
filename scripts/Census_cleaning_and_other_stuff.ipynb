```python
df2 = pd.read_csv('onlydata.csv', header=None)
senate_districts = df2[0].tolist()
senate_districts = senate_districts[:-1]
for i in range(len(senate_districts)):
    temp_str = senate_districts[i]
    senate_districts[i] = temp_str[4:]
census_senate = pd.read_csv('Census_Senate.csv')


census_senate['Geography'] = census_senate['Geography'].apply(lambda x: x.split("(")[0])
census_senate['Geography'] = census_senate['Geography'].apply(lambda x: x.split("District")[0])
    
index_list = []
for index, row in census_senate.iterrows():
    temp_str = row['Geography']
    if temp_str[:-1] not in senate_districts:
        index_list += [index]
census_senate = census_senate.drop(index_list, axis=0)
census_senate.to_csv('final_census_senate.csv')
print("final_census_senate is written")    
```
